```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# Duration, explicitly, and popularity of top tracks
# by Wuyi Zhang

import pandas as pd
import matplotlib.pyplot as plt
import collections
import json
```

```python
topTracks_Filenames = ['Top 10 Tracks of Drake.json',
                       'Top 10 Tracks of Justin Bieber.json',
                       'Top 10 Tracks of Rihanna.json',
                       'Top 10 Tracks of Twenty One Pilots.json',
                       'Top 10 Tracks of Kanye West.json']

topTracksForRelatedArtists_Filename = 'Top 10 Tracks of 25 Related Artists.json'

tracksInfo = []
```

```python
for file in topTracks_Filenames:
    with open(file) as dataFile:
        tracks = json.load(dataFile)['tracks']
        for track in tracks:
            tracksInfo.append(track)
```

```python
tracksInfoPd = pd.DataFrame(tracksInfo)
tracksInfoPd
```

```python
with open(topTracksForRelatedArtists_Filename) as dataFile:
    tracks = json.load(dataFile)
    for track in tracks:
        tracksInfo.append(track)
```

```python
tracksInfoPd = pd.DataFrame(tracksInfo)
tracksInfoPd
```

```python
tracksInfoPd['popularity'].describe()
```

```python
longestDuration = -999999999
mostPopularity = -999999999
for _, track in tracksInfoPd.iterrows():
#     print(track)
    if track['popularity'] > mostPopularity:
        mostPopularTracks = list()
        mostPopularTracks.append(track)
        mostPopularity = track['popularity']
    elif track['popularity'] == mostPopularity:
        mostPopularTracks.append(track)
    else:
        pass
    
    if track['duration_ms'] > longestDuration:
        longestDurationTracks = list()
        longestDurationTracks.append(track)
        longestDuration = track['duration_ms']
    elif track['duration_ms'] == longestDuration:
        longestDurationTracks.append(track)
    else:
        pass
```

```python
mostPopularTracksPd = pd.DataFrame(mostPopularTracks)
longestDurationTracksPd = pd.DataFrame(longestDurationTracks)
```
