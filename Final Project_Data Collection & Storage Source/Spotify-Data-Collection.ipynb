```python
#This script is written to extract data with Spotify API.
#NOTE: the tokens for Spotify API are likely to have expired. Running the script will yield to response 401 (access denied).
'''
1. Select top 5 most played artists;
2. Extract the related artists of the 5 each;
3. Extract top songs from each artist;
4. Extract song audio features for each song
by Yu Wang
'''
import requests
import json

top_five_artists = ['Drake','Justin Bieber','Rihanna','Twenty One Pilots','Kanye West']
top_five_artists_id = ['3TVXtAsR1Inumwj472S9r4','1uNFoZAHBGtllmzznpCI3s','5pKCCKE2ajJHZ9KAiaK11H','3YQKmKGau1PzlVlkL1iodx','5K4W6rqBFWDnAN6FQUkS6x']

#related artists
for i in range(5):
    raw_top_related = requests.get('https://api.spotify.com/v1/artists/'+top_five_artists_id[i]+'/related-artists'
)
    parsed_related_artists = json.loads(raw_top_related.text)
    fout = open('related_artist_'+str(i+1)+'_'+str(top_five_artists[i])+'.json','w')
    json.dump(parsed_related_artists,fout,indent=4)
    fout.close()
    
#top artists
raw_top_artists = requests.get('https://api.spotify.com/v1/artists?ids='+top_five_artists_id[0]+','+top_five_artists_id[1]+','+top_five_artists_id[2]+','+top_five_artists_id[3]+','+top_five_artists_id[4])

parsed_top_artists = json.loads(raw_top_artists.text)
fout = open('top_artists.json','w')
json.dump(parsed_top_artists,fout,indent=4)
fout.close()

#top tracks from top artists
for i in range(5):
    raw_top_tracks = requests.get('https://api.spotify.com/v1/artists/'+top_five_artists_id[i]+'/top-tracks?country=US')
    parsed_top_tracks = json.loads(raw_top_tracks.text)
    fout = open('top_tracks_'+str(i+1)+'_'+str(top_five_artists[i])+'.json','w')
    json.dump(parsed_top_tracks,fout,indent=4)
    fout.close()

#end of collecting data from API. API tokens may be expired
#extract track ID

file1 = json.load(open('top_tracks_1_Drake.json','r'))
file2 = json.load(open('top_tracks_2_Justin Bieber.json','r'))
file3 = json.load(open('top_tracks_3_Rihanna.json','r'))
file4 = json.load(open('top_tracks_4_Twenty One Pilots.json','r'))
file5 = json.load(open('top_tracks_5_Kanye West.json','r'))

file1_track_id = [track['id'] for track in file1['tracks']]
file2_track_id = [track['id'] for track in file2['tracks']]
file3_track_id = [track['id'] for track in file3['tracks']]
file4_track_id = [track['id'] for track in file4['tracks']]
file5_track_id = [track['id'] for track in file5['tracks']]

all_track_id = file1_track_id + file2_track_id + file3_track_id + file4_track_id +file5_track_id

#end of extracing track ID
#begin extracting track features
#NOTE: the tokens for Spotify API are likely to have expired. Running the script will yield to response 401 (access denied).

track_id_string = ','.join(all_track_id)

top_track_features_all_endpoint = 'https://api.spotify.com/v1/audio-features/?ids='+track_id_string
top_track_features_all_auth = {"Authorization":"Bearer BQAiqVrqh5siu6ENkkGq8Xg2f_i3fOh4syx7K2yPQ5hvapmpjJqhKm-_swJYae_-trM-7mPy5RRBXjZKTw1JXXq17zXqBG75D4J0Rm1HxxkFWO90Tm_NWlTR3HCyFFNlpTUqpBogoen1CVM"}
top_track_features_all = requests.get(top_track_features_all_endpoint,headers = top_track_features_all_auth)

parsed_top_track_features_all = json.loads(top_track_features_all.text)
fout = open('top_track_features_all.json','w')
json.dump(parsed_top_track_features_all,fout,indent=4)
fout.close()

#end of extracting song features
```
