```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# Most common genres for 2016 top artists
# by Wuyi Zhang

import pandas as pd
import matplotlib.pyplot as plt
import collections
import json
```

```python
artistsInfo_Filenames = ['Top 5 Artists.json', 
                         '5 Related Artist of Drake.json',
                         '5 Related Artist of Justin Bieber.json',
                         '5 Related Artist of Rihanna.json',
                         '5 Related Artist of Twenty One Pilots.json',
                         '5 Related Artist of Kanye West.json']
artistsInfo = []
```

```python
for file in artistsInfo_Filenames:
    with open(file) as dataFile:
#         print(json.load(dataFile)['artists'])
        artists = json.load(dataFile)['artists']
        for artist in artists:
            artistsInfo.append(artist)
artistsInfo
```

```python
artistsInfoDf = pd.DataFrame(artistsInfo)
artistsInfoDf
```

```python
frequencyDict = dict()
```

```python
for genreList in artistsInfoDf['genres']:
    for genre in genreList:
        if genre in frequencyDict.keys():
            frequencyDict[genre] += 1
        else:
            frequencyDict[genre] = 1
frequencyDict
```

```python
frequencyDictInversed = dict()
```

```python
for key in frequencyDict:
    if frequencyDict[key] in frequencyDictInversed.keys():
        frequencyDictInversed[frequencyDict[key]].append(key)
    else:
        frequencyDictInversed[frequencyDict[key]] = list()
        frequencyDictInversed[frequencyDict[key]].append(key)
frequencyDictInversed
```

```python
y, x = zip(*sorted(frequencyDictInversed.items()))
```
